```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
user_info = pd.read_csv('user_info_preprocess.csv', encoding='utf-8')
print(user_info.count())
user_info.head()
```

```python
train_data_listinfo = pd.read_csv('train_data_listing_info.csv')
print(train_data_listinfo.count())
train_data_listinfo.head()
```

```python
def join_userinfo(row):
#     print(row)
    max_ts = row['auditing_ts']
    userid = row['user_id']
    matchinfo = user_info.loc[user_info['user_id'] == userid]
    if len(matchinfo) > 1:
        matchinfo = matchinfo.sort_values('insert_ts', ascending=False)
        for i,m in matchinfo.iterrows():
            if m['insert_ts'] < max_ts:
                row['reg_mon'] = m['reg_mon']
                row['gender'] = m['gender']
                row['age'] = m['age']
                row['cell_province'] = m['cell_province']
                row['id_province'] = m['id_province']
                row['id_city'] = m['id_city']
                row['insertdate'] = m['insertdate']
                row['insert_ts'] = m['insert_ts']
                break
    else:
        m = matchinfo.iloc[0]
        row['reg_mon'] = m['reg_mon']
        row['gender'] = m['gender']
        row['age'] = m['age']
        row['cell_province'] = m['cell_province']
        row['id_province'] = m['id_province']
        row['id_city'] = m['id_city']
        row['insertdate'] = m['insertdate']
        row['insert_ts'] = m['insert_ts']
    return row
train_data_listinfo_userinfo = train_data_listinfo.apply(join_userinfo, axis=1)
```

```python
train_data_listinfo_userinfo.count()
```

```python
train_data_listinfo_userinfo.to_csv('train_data_listinfo_userinfo.csv', index=False, encoding='utf-8')
```
